Cargar, mostrar y guardar imágenes

```python
# Instalar OpenCV si es necesario (en Colab ya está instalado)
#!pip install opencv-python

import cv2
import numpy as np
from matplotlib import pyplot as plt
import urllib.request
from google.colab.patches import cv2_imshow
import os
```

```python
# Descargar una imagen de ejemplo para el tutorial
print("\nDescargando imagen de ejemplo...")
url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg"
filename = "lena.jpg"
urllib.request.urlretrieve(url, filename)
```

```python
# 1.- Cargar una imagen

# Cragar umagen en color
imagen_color = cv2.imread(filename)

# Cargar imagen en escala de grises
imagen_gris = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
```

```python
# 2. MOSTRAR PROPIEDADES DE LA IMAGEN
print("\n3. Propiedades de la imagen:")
print(f"   - Forma (color): {imagen_color.shape}")  # (alto, ancho, canales)
print(f"   - Forma (gris): {imagen_gris.shape}")    # (alto, ancho)
print(f"   - Tipo de dato: {imagen_color.dtype}")
print(f"   - Valor máximo: {np.max(imagen_color)}")
print(f"   - Valor mínimo: {np.min(imagen_color)}")
```

```python
# 3. MOSTRAR IMÁGENES
print("\n3. Mostrando imágenes...")

# Método 1: Usando cv2_imshow (recomendado para Colab)
print("\n📸 Imagen en color (OpenCV - BGR):")
cv2_imshow(imagen_color)
```

```python
print("\n📸 Imagen en escala de grises:")
cv2_imshow(imagen_gris)
```

```python
# Método 2: Usando matplotlib (mejor para mostrar imágenes RGB)
print("\n🎨 Imagen convertida a RGB (Matplotlib):")

# Convertir BGR a RGB para matplotlib
imagen_rgb = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2RGB)
```

```python
# Crear mosaico
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(imagen_rgb)
plt.title("Imagen Color (RGB)")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(imagen_gris, cmap='gray')
plt.title("Escala de Grises")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(imagen_color, cv2.COLOR_BGR2HSV))
plt.title("Espacio HSV")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
print("\n⚫ Imagen en escala Lab:")
imagen_lab = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2LAB)
cv2_imshow(imagen_lab)
```

```python
# 4. Guardar imagenes
print("\n5. Guardando imágenes...")

# Guardar imagen en escala de grises
nombre_gris = "lena_gris.jpg"
cv2.imwrite(nombre_gris, imagen_gris)
```

```python
# Guardar imagen en color
nombre_color = "lena_color_guardada.jpg"
cv2.imwrite(nombre_color, imagen_color)
```

```python
# Crear y guardar una imagen modificada
imagen_modificada = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2GRAY)
imagen_modificada = cv2.GaussianBlur(imagen_modificada, (15, 15), 0)
nombre_modificada = "lena_modificada.jpg"
cv2.imwrite(nombre_modificada, imagen_modificada)
```

```python
# 5. En resumen
print("📝 Resumen de funciones utilizadas:")
print("   - cv2.imread(): Cargar imagen")
print("   - cv2_imshow(): Mostrar imagen en Colab")
print("   - cv2.imwrite(): Guardar imagen")
print("   - cv2.cvtColor(): Convertir entre espacios de color")
print("   - cv2.GaussianBlur(): Aplicar filtro gaussiano")
```

```python
# Guardar imagen en png
nombre_color = "lena_color_guardada.png"
cv2.imwrite(nombre_color, imagen_color)
```
